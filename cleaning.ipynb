```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('autos.csv')
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
df['dateCrawled'] = pd.to_datetime(df['dateCrawled'], errors='coerce')
df['dateCreated'] = pd.to_datetime(df['dateCreated'], errors='coerce')
df['lastSeen'] = pd.to_datetime(df['lastSeen'], errors='coerce')

df['dateCrawled'] = df['dateCrawled'].dt.date
df['dateCreated'] = df['dateCreated'].dt.date
df['lastSeen'] = df['lastSeen'].dt.date

df
```

```python
df ['name'] = df['name'].replace('_', ' ', regex=True)
df
```

```python
df = df[df['yearOfRegistration'] > 1900]
df
```

```python
df = df.drop(['postalCode', 'nrOfPictures', 'index'], axis=1)
```

```python
df['notRepairedDamage'] = df['notRepairedDamage'].replace('nein', 'no')
df['notRepairedDamage'] = df['notRepairedDamage'].replace('ja', 'yes')

df['offerType'] = df['offerType'].replace('angebot', 'offer')
df['offerType'] = df['offerType'].replace('gesuch', 'request')
df['vehicleType'] = df['vehicleType'].replace('kleinwagen', 'hatchback')
df['vehicleType'] = df['vehicleType'].replace('limousine', 'sedan')
df['vehicleType'] = df['vehicleType'].replace('kombi', 'station wagon')
df['vehicleType'] = df['vehicleType'].replace('andere', 'other')

df['seller'] = df['seller'].replace('privat', 'private')
df['seller'] = df['seller'].replace('gewerblich', 'commercial')
df['seller'] = df['seller'].replace('Angebot', 'offer')

df['fuelType'] = df['fuelType'].replace('benzin', 'gasoline')
df['fuelType'] = df['fuelType'].replace('elektro', 'electric')
df['fuelType'] = df['fuelType'].replace('andere', 'other')

df['gearbox'] = df['gearbox'].replace('manuell', 'manual')
df['gearbox'] = df['gearbox'].replace('automatik', 'automatic')

df
```

```python
df = df.dropna(subset=['notRepairedDamage', 'vehicleType', 'powerPS', 'fuelType', 'model', 'gearbox'])
df
```

```python
valid_fuels = ['gasoline', 'diesel', 'lpg', 'electric', 'other','cng']
df = df[df['fuelType'].isin(valid_fuels)]
valid_gearboxes = ['manual', 'automatic']
df = df[df['gearbox'].isin(valid_gearboxes)]
non_valid_brands = ['ja','nein']
df = df[~df['brand'].isin(non_valid_brands)]
no_valid_vechile_types = ['other']
df = df[~df['vehicleType'].isin(no_valid_vechile_types)]
```

```python
df = df[df['offerType'] != 'request']
df
df.drop('offerType', axis=1, inplace=True)
df
```

```python
df = df.drop(columns=['seller'])
```

```python
df
```

```python
df['car_age'] = 2016 - df['yearOfRegistration']
df
```

```python
df.loc[df['car_age'] == 0, 'car_age'] = 1
```

```python
df['price_in_thousands'] = df['price'] / 1000
df.drop('price', axis=1, inplace=True)
df
```

```python
df['kilometer']  = df['kilometer'] / 1000
df['powerPS']  = df['powerPS'].astype(float)
df['car_age']  = df['car_age'].astype(float)
df
```

```python
df['price_in_thousands'] = df['price_in_thousands'].fillna(df.groupby('brand')['price_in_thousands'].transform('mean'))
df
```

```python
df.isnull().sum()
```

```python
df.loc[df['price_in_thousands'] == 0 , 'price_in_thousands'] = df.groupby('brand')['price_in_thousands'].transform('mean')
df
```

```python
df.to_csv('cleaned_autos.csv', index=False)
```
